```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings 
warnings.filterwarnings('ignore')
import folium
from folium.plugins import HeatMap
```

```python
df = pd.read_csv('datasets\\Chicago_Crimes.csv')
df
```

```python
df.info()
```

```python
df['ID'] = df['ID'].astype('category')
df['Case Number'] = df['Case Number'].astype('category')
df['Date'] = df['Date'].astype('category')
df['Block'] = df['Block'].astype('category')
df['IUCR'] = df['IUCR'].astype('category')
df['Year'] = df['Year'].astype('category')
df['Location Description'] = df['Location Description'].astype('category')
df['District'] = df['District'].astype('category')
df['Ward'] = df['Ward'].astype('category')
df['FBI Code'] = df['FBI Code'].astype('category')
```

```python
df.columns
```

```python
df['Primary Type'].value_counts()
```

```python
THEFT = df[df['Primary Type']=='THEFT']
BATTERY = df[df['Primary Type']=='BATTERY']
CRIMINALDAMAGE = df[df['Primary Type']=='CRIMINAL DAMAGE']
ASSAULT = df[df['Primary Type']=='ASSAULT']
MOTORVEHICLETHEFT = df[df['Primary Type']=='MOTOR VEHICLE THEFT']
OTHEROFFENSE = df[df['Primary Type']=='OTHER OFFENSE']
DECEPTIVEPRACTICE = df[df['Primary Type']=='DECEPTIVE PRACTICE']
BURGLARY = df[df['Primary Type']=='BURGLARY']
ROBBERY = df[df['Primary Type']=='ROBBERY']
WEAPONSVIOLATION = df[df['Primary Type']=='WEAPONS VIOLATION']
NARCOTICS = df[df['Primary Type']=='NARCOTICS']
CRIMINALTRESPASS = df[df['Primary Type']=='CRIMINAL TRESPASS']
OFFENSEINVOLVINGCHILDREN = df[df['Primary Type']=='OFFENSE INVOLVING CHILDREN']
CRIMINALSEXUALASSAULT = df[df['Primary Type']=='CRIMINAL SEXUAL ASSAULT']
SEXOFFENSE = df[df['Primary Type']=='SEX OFFENSE']
PUBLICPEACEVIOLATION = df[df['Primary Type']=='PUBLIC PEACE VIOLATION']
INTERFERENCEWITHPUBLIC_OFFICER = df[df['Primary Type']=='INTERFERENCE WITH PUBLIC OFFICER']
HOMICIDE = df[df['Primary Type']=='HOMICIDE']
STALKING = df[df['Primary Type']=='STALKING']
ARSON = df[df['Primary Type']=='ARSON']
PROSTITUTION = df[df['Primary Type']=='PROSTITUTION']
CONCEALEDCARRYLICENSEVIOLATION = df[df['Primary Type']=='CONCEALED CARRY LICENSE VIOLATION']
LIQUORLAWVIOLATION = df[df['Primary Type']=='LIQUOR LAW VIOLATION']
INTIMIDATION = df[df['Primary Type']=='INTIMIDATION']
KIDNAPPING = df[df['Primary Type']=='KIDNAPPING']
OBSCENITY = df[df['Primary Type']=='OBSCENITY']
HUMANTRAFFICKING = df[df['Primary Type']=='HUMAN TRAFFICKING']
GAMBLING = df[df['Primary Type']=='GAMBLING']
PUBLICINDECENCY = df[df['Primary Type']=='PUBLIC INDECENCY']
OTHER_NARCOTICVIOLATION = df[df['Primary Type']=='OTHER NARCOTIC VIOLATION']
NONCRIMINAL = df[df['Primary Type']=='NON-CRIMINAL']
```

```python
THEFTloc = THEFT[THEFT['Location Description'] == "APARTMENT"]
THEFTlocarrest = THEFTloc[THEFTloc['Arrest'] == 1]
THEFTlocarrest
```

```python
heatmap1 = THEFTlocarrest.groupby(['Latitude','Longitude']).size().reset_index(name="incident_count")
heatmap1data = heatmap1[['Latitude','Longitude','incident_count']].values.tolist()
map1 = folium.Map(location = [30,40], zoom_start=4)
HeatMap(heatmap1data).add_to(map1)
```

```python
map1.save('chicago_crimes.html')
print("Data Successfully saved to chicago_crimes.html")
map1
```
