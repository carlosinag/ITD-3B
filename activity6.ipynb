```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings 
warnings.filterwarnings('ignore')
```

```python
import folium
from folium.plugins import HeatMap
```

```python


# Load the dataset with the correct encoding
df = pd.read_csv('datasets\\Global_terrorist.csv', encoding='ISO-8859-1')

```

```python
df
```

```python
ph = df[df['country_txt'] == "Philippines"]
```

```python
mideastnafrica = df[df['region_txt']== "Middle East & North Africa"]
mideastnafrica2000up = mideastnafrica[mideastnafrica['iyear'] > 2000]
mideastnafrica2000upattacktype = mideastnafrica2000up[mideastnafrica2000up['attacktype1_txt']== 'Bombing/Explosion']
```

```python
heatmap1 = mideastnafrica2000upattacktype.groupby(['latitude','longitude']).size().reset_index(name="incident_count")
heatmap1
```

```python
heatmap1data = heatmap1[['latitude','longitude','incident_count']].values.tolist()
map1 = folium.Map(location = [30,40], zoom_start=4)
```

```python
HeatMap(heatmap1data).add_to(map1)
```

```python
map1.save('terrorism_bombing.html')
print("Data Successfully saved to terrorism_bombing.html")
```

```python
map1
```
